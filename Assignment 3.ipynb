# Question 1. <code>euro_jackpot()</code>

EuroJackpot prizes are amongst the largest in Europe, second only to EuroMillions in terms of the total money paid out. In EuroJackpot, you simply select **five** regular numbers from 1 to 50 and then the **two** additional Euro numbers from 1 to 10. You are trying to predict as many numbers as possible and will win the jackpot if you get all seven correctly.



For more information about EuroJackpot, please click [here](https://www.eurojackpot.org/en/bet-on-eurojackpot/). 

Please write a function <code>euro_jackpot()</code>, which generates a random selection of **seven** numbers for a EuroJackpot ticket. Among the seven numbers to be generated:
 - The first five numbers (*i.e.*, regular numbers) are randomly selected from 1 to 50 (both inclusive), and do not contain any duplicates.
 - The other two numbers (*i.e.*, Euro numbers) are randomly selected from 1 to 10 (both inclusive), and do not contain any duplicates.

Note that the two Euro numbers are drawn from a separate ball machine, so it is possible to have the same number as both a regular number and a Euro number. Your function should first display the five regular numbers in ascending order, and then the two Euro numbers in ascending order.

```python
## Your function starts from here:

def euro_jackpot():
    
## To get First five random numbers between 1-50 (both inclusive) in ascending order
    first_five = sorted(np.random.randint(1, 51, 5))
    
## To get Last two random numbers between 1-10 (both inclusive) in ascending order
    last_two = sorted(np.random.randint(1, 11, 2))

## To get the final result in desired format
    print(', '.join(map(str, first_five)) + " - " + ', '.join(map(str, last_two)))
```

# Question 2.

Four customers visited a chocolate store, each of which bought chocolates out of three brands (A, B and C):
 - Lucas bought 100 grams of brand A, 175 grams of brand B and 210 grams of brand C.
 - Elina bought 90 grams of A, 160 grams of B and 150 grams of C.
 - Veera bought 200 grams of A, 50 grams of B and 100 grams of C.
 - Mia bought 120 grams of A and 310 grams of C.
 
The price for each brand is as follows:
 - A costs €2.98 per 100 grams
 - B costs €3.90 per 100 grams
 - C costs €1.99 per 100 grams

Please calculate how much each person had to pay.

NOTE: You can ONLY use element-wise operations and universal functions from <code>NumPy</code>, and you are NOT allowed to use any loops (e.g. <code>for</code> or <code>while</code>).

```python
## Your program starts from here:

## Lucas, Elina, Veera and Mia bought the amounts of chocolates out of (A, B, C) brands respectively 
Lucas = np.array([100, 175, 210])  
Elina = np.array([90, 160, 150])
Veera = np.array([200, 50, 100])
Mia = np.array([120, 0, 310])

## Cost per gram of chocolates of A, B, C brands respectively
CpG = np.array([2.98 / 100, 3.90 / 100, 1.99 / 100]) 

## Calculating Total cost of each person behind all 03 brands (element wise operation)
L = sum(Lucas * CpG) 
E = sum(Elina * CpG)
V = sum(Veera * CpG)
M = sum(Mia * CpG)

## Printing final result in desired 02 digit format after the decimal point 
print('Lucas had to pay €' + '%.2f'% L)
print('Elina had to pay €' + '%.2f'% E)
print('Veera had to pay €' + '%.2f'% V)
print('Mia had to pay €' + '%.2f'% M)
```
